```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from pandas import DataFrame
from sklearn import preprocessing

# Reading data from the csv file containing the dataset and storing it in a numpy array
```

```python
dataset = pd.read_csv('movies.csv', encoding = "ISO-8859-1")
dimensions = np.array(dataset)
list_key_dim = []
#all_dimensions = dimensions, key_dimensions_list = list_key_dim
#Removing the datapoints where the US Box Office Revenue is zero, 
#extracting the key dimensions and appending them to a numpy array
for i in range(len(dimensions)):
    if dimensions[i][0]!=0:
        list_key_dim.append((dimensions[i][0], dimensions[i][5], dimensions[i][10], dimensions[i][12]))
        
print('The number of datapoints is now:', len(list_key_dim))
```

```python
df = DataFrame (list_key_dim)
print (df)
#list_key_dim.append((dimensions[i][0], dimensions[i][5], dimensions[i][10]))
```

```python
df.head(10)
```

```python
df = DataFrame (list_key_dim, columns=['budget','gross','score','votes'])
print (df)
```

```python
df.shape
```

```python
#we are going to do standardisation and going to use PCA to reduce the 3 dimensional space into one dimensional space
#rescaling feature vectors to all have the same scale
# X is feature vectors
X = df[['budget', 'gross', 'score', 'votes']]
X.head(9)
```

```python
#Y = df[['votes']]
#Y.head(10)
```

```python
dim = np.array(X)
#Scaling the dataset
scaler = StandardScaler()
dim_sc = scaler.fit_transform(dim)
```

```python
#Phase 1 - Creating three 2D scatterplots of the key dimensions that I have selected

plt.scatter(dim_sc[:, 0], dim_sc[:, 1], color = 'red', s=0.5)
plt.title('Budget and Gross')
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.show()
```

```python
plt.scatter(dim_sc[:, 0], dim_sc[:, 2], color = 'blue', s=0.5)
plt.title('Budget and Score(IMDB Rating)')
plt.xlabel('Budget')
plt.ylabel('Score')
plt.show()
```

```python
plt.scatter(dim_sc[:, 2], dim_sc[:, 1], color = 'green', s=0.5)
plt.title('Score(IMDB Rating) and Gross(US Box Office Revenue)')
plt.xlabel('Score')
plt.ylabel('Gross')
plt.show()
```

```python
plt.scatter(dim_sc[:, 3], dim_sc[:, 1], color = 'orange', s=0.5)
plt.title('Votes and Gross(US box office revenue)')
plt.xlabel('Votes')
plt.ylabel('Gross')
plt.show()
```

```python
#Phase 2 - Creating three 2D numpy arrays of the 2D combinations of the 3D data

dim_sc1 = []
dim_sc2 = []
dim_sc3 = []
dim_sc4 = []

for i in range(len(dim_sc)):
    dim_sc1.append((dim_sc[i][0], dim_sc[i][1]))
    dim_sc2.append((dim_sc[i][0], dim_sc[i][2]))
    dim_sc3.append((dim_sc[i][2], dim_sc[i][1]))
    dim_sc4.append((dim_sc[i][3], dim_sc[i][1]))
    
dim_sc1 = np.array(dim_sc1)
dim_sc2 = np.array(dim_sc2)
dim_sc3 = np.array(dim_sc3)
dim_sc4 = np.array(dim_sc4)
```

```python
#Performing the PCA
pca = PCA(n_components = 1)
dim_pca_1 = pca.fit_transform(dim_sc1)
dim_pca_2 = pca.fit_transform(dim_sc2)
dim_pca_3 = pca.fit_transform(dim_sc3)
dim_pca_4 = pca.fit_transform(dim_sc4)
```

```python
# Phase 3 - Creating three 1D scatterplots of the key dimensions
plt.hlines(1, min(min(dim_pca_1)), max(max(dim_pca_1)))
y = np.ones(np.shape(dim_pca_1))
plt.plot(dim_pca_1, y, '|', ms = 10, color = 'red')
plt.title('Budget and Gross')
plt.axis('off')
plt.show()
```

```python
plt.hlines(1, min(min(dim_pca_2)), max(max(dim_pca_2)))
y = np.ones(np.shape(dim_pca_2))
plt.plot(dim_pca_2, y, '|', ms = 10, color = 'blue')
plt.title('Budget and Score')
plt.axis('off')
plt.show()
```

```python
plt.hlines(1, min(min(dim_pca_3)), max(max(dim_pca_3)))
y = np.ones(np.shape(dim_pca_3))
plt.plot(dim_pca_3, y, '|', ms = 10, color = 'green')
plt.title('Score and Gross')
plt.axis('off')
plt.show()
```

```python
plt.hlines(1, min(min(dim_pca_4)), max(max(dim_pca_4)))
y = np.ones(np.shape(dim_pca_4))
plt.plot(dim_pca_4, y, '|', ms = 10, color = 'orange')
plt.title('Votes and Gross')
plt.axis('off')
plt.show()
```

```python
#data preparation to create the Scree Plots of the PCA
screeplot = PCA(n_components = 2)

screeplot.fit_transform(dim_sc1)
screeplot1 = screeplot.explained_variance_ratio_
screeplot1 = np.array(screeplot1).cumsum()

screeplot.fit_transform(dim_sc2)
screeplot2 = screeplot.explained_variance_ratio_
screeplot2 = np.array(screeplot2).cumsum()

screeplot.fit_transform(dim_sc3)
screeplot3 = screeplot.explained_variance_ratio_
screeplot3 = np.array(screeplot3).cumsum()

screeplot.fit_transform(dim_sc4)
screeplot4 = screeplot.explained_variance_ratio_
screeplot4 = np.array(screeplot4).cumsum()

#Creating the Scree Plots of the PCA
plt.bar([1,2], screeplot1, color = 'red')
plt.xticks([1,2])
plt.title('Scree Plot for Budget and Gross')
plt.xlabel('Principal Component')
plt.ylabel('Cumulative Proportion of Variance Explained')
plt.show()

plt.bar([1,2], screeplot2, color = 'blue')
plt.xticks([1,2])
plt.title('Scree Plot for Budget and Score')
plt.xlabel('Principal Component')
plt.ylabel('Cumulative Proportion of Variance Explained')
plt.show()

plt.bar([1,2], screeplot3, color = 'green')
plt.xticks([1,2])
plt.title('Scree Plot for Score and Gross')
plt.xlabel('Principal Component')
plt.ylabel('Cumulative Proportion of Variance Explained')
plt.show()

plt.bar([1,2], screeplot4, color = 'orange')
plt.xticks([1,2])
plt.title('Scree Plot for Votes and Gross')
plt.xlabel('Principal Component')
plt.ylabel('Cumulative Proportion of Variance Explained')
plt.show()
```
