# Data Visualization

```python
#grouping of this data on the basis of category
grp = data.groupby('genre')
x = grp['votes'].agg(np.mean)
y = grp['runtime'].agg(np.mean)
z = grp['budget'].agg(np.mean)
a = grp['score'].agg(np.mean)
b = grp['gross'].agg(np.mean)
print(x)
print(y)
print(z)
print(a)
print(b)

#An app with a name A, B, N, M -- that belongs to Art and Design so it is grouping the apps together, the game apps together 
#so its grouping the apps together and storing it in the variable grp

#Now from this grp variable, extracting the rating column and aggregating it with mean
#for eg if in art and design we have 13 apps -{4,3,4.3,5,...}<--mean of this ratings and storing it in my variable x
```

```python
#the wrote the title which is Genre wise gross the x labelled to genre
plt.figure(figsize=(16,5)) # length to be 12 and height to be 5
plt.plot(x, 'ro', color='r')#by default the colour is red we could use blue, green
plt.xticks(rotation=90)# then showed the list of nos which I dont want so I did a plt .show
plt.title('Genre wise Votes')
plt.xlabel('Genre-->')
plt.ylabel('Votes-->')
plt.show()
```

```python
#the wrote the title which is Genre wise gross the x labelled to genre
plt.figure(figsize=(16,5)) # length to be 12 and height to be 5
plt.plot(y, 'r--', color='b')#by default the colour is red we could use blue, green
plt.xticks(rotation=90)# then showed the list of nos which I dont want so I did a plt .show
plt.title('Genre wise Runtime')
plt.xlabel('Genre-->')
plt.ylabel('Runtime-->')
plt.show()
```

```python
#the wrote the title which is Genre wise gross the x labelled to genre
plt.figure(figsize=(16,5)) # length to be 12 and height to be 5
plt.plot(z, 'bs', color='g')#by default the colour is red we could use blue, green
plt.xticks(rotation=90)# then showed the list of nos which I dont want so I did a plt .show
plt.title('Genre wise Budget')
plt.xlabel('Genre-->')
plt.ylabel('Budget-->')
plt.show()
```

```python
#the wrote the title which is Genre wise gross the x labelled to genre
plt.figure(figsize=(16,5)) # length to be 12 and height to be 5
plt.plot(a, 'g^', color='r')#by default the colour is red we could use blue, green
plt.xticks(rotation=90)# then showed the list of nos which I dont want so I did a plt .show
plt.title('Genre wise Score')
plt.xlabel('Genre-->')
plt.ylabel('Score-->')
plt.show()
```

```python
#the wrote the title which is Genre wise gross the x labelled to genre
plt.figure(figsize=(16,5)) # length to be 12 and height to be 5
plt.plot(b, 'r--', color='b')#by default the colour is red we could use blue, green
plt.xticks(rotation=90)# then showed the list of nos which I dont want so I did a plt .show
plt.title('Genre wise Gross')
plt.xlabel('Genre-->')
plt.ylabel('Gross-->')
plt.show()
```
